```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
```

## Load audio files

```python
bad_file = "media/bad_bearing.wav"
good_file = "media/good_bearing.wav"
```

```python
ipd.Audio('media/bad_bearing.wav') # load a local WAV file
```

```python
ipd.Audio('media/good_bearing.wav') # load a local WAV file
```

## Mel filter banks

```python
# load audio files with librosa
bad, sr = librosa.load(bad_file)
```

```python
filter_banks = librosa.filters.mel(n_fft=256, sr=sr, n_mels=64)
```

```python
filter_banks.shape
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(filter_banks, sr=sr, x_axis="linear")
plt.colorbar(format="%+2.f")
plt.show()
```

```python
# load audio files with librosa
good, sr = librosa.load(good_file)
```

```python
filter_banks = librosa.filters.mel(n_fft=256, sr=sr, n_mels=64)
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(filter_banks, sr=sr, x_axis="linear")
plt.colorbar(format="%+2.f")
plt.show()
```

## Extracting Mel Spectrogram

```python
mel_spectrogram = librosa.feature.melspectrogram(bad, sr=sr, n_fft=256, hop_length=32, n_mels=64)
```

```python
mel_spectrogram.shape
```

```python
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
```

```python
log_mel_spectrogram.shape
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(log_mel_spectrogram, x_axis="time", y_axis="mel", sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
mel_spectrogram = librosa.feature.melspectrogram(good, sr=sr, n_fft=256, hop_length=32, n_mels=64)
```

```python
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(log_mel_spectrogram, x_axis="time", y_axis="mel", sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```
