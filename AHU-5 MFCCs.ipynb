```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

## Loading audio files

```python
bad_file = "media/bad_bearing.wav"
good_file = "media/good_bearing.wav"
```

```python
# load audio files with librosa
bad, sr = librosa.load(bad_file)
good, _ = librosa.load(good_file)
```

## Extracting MFCCs

```python
mfccs = librosa.feature.mfcc(y=bad, n_mfcc=88, sr=sr)
```

```python
mfccs.shape
```

## Visualising MFCCs

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(mfccs, x_axis="time", sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
mfccs = librosa.feature.mfcc(y=good, n_mfcc=88, sr=sr)
```

```python
mfccs.shape
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(mfccs, x_axis="time", sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

## Computing first / second MFCCs derivatives

```python
mfccs = librosa.feature.mfcc(y=bad, n_mfcc=88, sr=sr)
```

```python
delta_mfccs = librosa.feature.delta(mfccs)
```

```python
delta2_mfccs = librosa.feature.delta(mfccs, order=2)
```

```python
delta_mfccs.shape
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta_mfccs, x_axis="time", sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta2_mfccs, x_axis="time", sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
mfccs = librosa.feature.mfcc(y=good, n_mfcc=88, sr=sr)
```

```python
delta_mfccs = librosa.feature.delta(mfccs)
```

```python
delta2_mfccs = librosa.feature.delta(mfccs, order=2)
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta_mfccs, x_axis="time", sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta2_mfccs, x_axis="time", sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```
