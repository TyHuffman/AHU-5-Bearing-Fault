```python
def plot_magnitude_spectrum(signal, sr, title, f_ratio=10):
    X = np.fft.fft(signal)
    X_mag = np.absolute(X)
    
    plt.figure(figsize=(18, 5))
    
    f = np.linspace(0, sr, len(X_mag))
    f_bins = int(len(X_mag)*f_ratio)  
    
    plt.plot(f[:f_bins], X_mag[:f_bins])
    plt.xlabel('Frequency (Hz)')
    plt.title(title)
```

```python
plot_magnitude_spectrum(bad, sr, "Bad Bearing", 0.12)
```

```python
plot_magnitude_spectrum(good, sr, "Good Bearing", 0.12)
```
