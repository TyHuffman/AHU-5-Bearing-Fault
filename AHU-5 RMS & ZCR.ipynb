```python
import matplotlib.pyplot as plt
import numpy as np
import librosa
import librosa.display
import IPython.display as ipd
```

## Import Audio

```python
bad_file = "media/bad_bearing.wav"
good_file = "media/good_bearing.wav"
```

```python
from IPython.display import Audio 
Audio('media/bad_bearing.wav', autoplay=False)
```

```python
from IPython.display import Audio 
Audio('media/good_bearing.wav', autoplay=False)
```

```python
# load audio files with librosa
bad, sr = librosa.load(bad_file)
good, _ = librosa.load(good_file)
```

```python
FRAME_SIZE = 200
HOP_LENGTH = 64
```

## RMS Root Mean Square

```python
rms_bad = librosa.feature.rms(bad, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
rms_good = librosa.feature.rms(good, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
```

## Visualise RMSE + waveform

```python
frames = range(len(rms_bad))
t = librosa.frames_to_time(frames, hop_length=HOP_LENGTH)
```

```python
# rms energy is graphed in red

plt.figure(figsize=(15, 17))

ax = plt.subplot(2, 1, 1)
librosa.display.waveplot(bad, alpha=0.5)
plt.plot(t, rms_bad, color="r")
plt.ylim((-1, 1))
plt.title("Bad")

plt.subplot(2, 1, 2)
librosa.display.waveplot(good, alpha=0.5)
plt.plot(t, rms_good, color="r")
plt.ylim((-1, 1))
plt.title("Good")

plt.show()
```

## Zero-crossing rate

```python
zcr_bad = librosa.feature.zero_crossing_rate(bad, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
zcr_good = librosa.feature.zero_crossing_rate(good, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
```

```python
plt.figure(figsize=(15, 10))

plt.plot(t, zcr_bad, color="r")
plt.plot(t, zcr_good, color="g")
plt.ylim(0, 0.3)
plt.show()
```
